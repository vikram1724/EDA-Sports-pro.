```python
import cmocean
import os
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
mpl.style.use('ggplot')
file_directory = os.path.abspath(os.path.join(os.path.dirname(os.getcwd()), '..')) + '\\'
report_directory = file_directory + 'predictive/report/'
```

```python
rmse_report = pd.read_csv(report_directory + 'rmse_report_fs_meta_stage.csv', index_col=0)
rsquare_report = pd.read_csv(report_directory + 'rsquare_report_tuned.csv', index_col=0)
meta_report = pd.read_csv(report_directory + 'meta_report_tuned.csv', index_col=0)
```

```python
# Get performance of all models
mpl.rcParams['figure.figsize'] = (20.0, 15.0)
value_df = rmse_report.dropna().astype(float).iloc[:, :]
sns.heatmap(value_df, linewidths=.1, 
            annot=True, fmt='.3f', cmap=cmocean.cm.curl, cbar=False, annot_kws={"size": 15}) 
```

```python
mpl.rcParams['figure.figsize'] = (20.0, 15.0)
sns.heatmap(meta_report.iloc[:, :-1], linewidths=.1, 
            annot=True, fmt='.3f', cmap=cmocean.cm.curl, cbar=False, annot_kws={"size": 15}) 
```
